```python
import pandas as pd
import numpy as np
import requests
header = {'User-Agent': "mr.muffin235@gmail.com"}
companyTickers = requests.get("https://www.sec.gov/files/company_tickers.json", headers=header)
T_list = companyTickers.json().values()
pd.set_option('display.max_columns', None)
```

```python
#account = ['AssetsCurrent', 'Liabilities']
accL = 'Liabilities'
accA = 'AssetsCurrent'
year = '2022'
```

```python
companyFramesL = requests.get((f'https://data.sec.gov/api/xbrl/frames/us-gaap/{accL}/USD/CY{year}Q4I.json'),
                                 headers=header)
```

```python
companyFramesA = requests.get((f'https://data.sec.gov/api/xbrl/frames/us-gaap/{accA}/USD/CY{year}Q4I.json'),
                                 headers=header)
```

```python
assetsDf = pd.DataFrame(companyFramesA.json()['data'])
```

```python
liabilityDf = pd.DataFrame(companyFramesL.json()['data'])
```

```python
liabilityDf.rename(columns={'val': 'Liabilities'}, inplace = True)
```

```python
assetsDf.rename(columns={'val': 'Assets'}, inplace = True)
```

```python
merged_df = pd.merge(liabilityDf, assetsDf, on='cik')
```

```python
merged_df['Acid_Ratio'] = merged_df['Assets'] / merged_df['Liabilities']
```

```python
merged_df.drop(axis = 1, columns=['accn_x', 'accn_y','loc_y', 'end_y', 'entityName_y'], inplace=True)
```

```python
merged_df.rename(columns={'loc_x': 'loc', 'end_x': 'end','entityName_x': 'entityName'}, inplace = True)
```

```python
merged_df.sort_values('Acid_Ratio',ascending = False, inplace = True)
```
